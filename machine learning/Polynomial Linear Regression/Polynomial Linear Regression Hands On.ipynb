```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Position_Salaries.csv")
data
```

## Spliiting the data

```python
x = data.iloc[:,1:-1].values
y = data.iloc[:,-1].values
```

```python
x
```

```python
y
```

## Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x,y)
```

## Visualise the LR Model

```python
plt.scatter(x,y,color = "red")
plt.plot(x,lin_reg.predict(x),color = "blue")
plt.xlabel("Input")
plt.ylabel("predicion")
plt.show()
```

## polynomial Linear Regression

```python
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree = 2)
x_poly = poly_reg.fit_transform(x)
```

```python
x_poly
```

```python
poly_linear_reg = LinearRegression()
poly_linear_reg.fit(x_poly,y)
```

## visualising polynomial Linear Regression

```python
plt.scatter(x,y,color = "red")
plt.plot(x,poly_linear_reg.predict(poly_reg.fit_transform(x)), color = "blue")
plt.xlabel("Input")
plt.ylabel("Prediction")
plt.show()
```
